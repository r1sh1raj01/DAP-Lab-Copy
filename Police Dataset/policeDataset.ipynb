## Police Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
df_police=pd.read_csv('Police Data.csv')
```

```python
df_police.head()
```

```python
# Q. 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.
df_police.dropna(axis=1, how='all', inplace=True)
df_police.head()
```

```python
# Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ? 
speeding_data = df_police[df_police['violation'] == 'Speeding']
speeding_counts_by_gender = speeding_data['driver_gender'].value_counts()
speeding_counts_by_gender
```

# Observation : Male were stopped more often

```python
# Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ?
search_freq_by_gender = df_police.groupby('driver_gender')['search_conducted'].mean()
search_freq_by_gender.head()
```

# boservation :male got searched

```python
# Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?
# Mapping stop_duration to numerical values
mapping_duration = {'0-15 Min': 7.5, '16-30 Min': 23, '30+ Min': 45}
df_police['stop_duration_numeric'] = df_police['stop_duration'].map(mapping_duration)

# Calculating mean stop_duration
mean_stop_duration = df_police['stop_duration_numeric'].mean()
mean_stop_duration
```

```python
# Q. 5) Question ( Groupby , Describe ) - Compare the age distributions for each violation
age_stats_by_violation = df_police.groupby('violation')['driver_age'].describe()
age_stats_by_violation.head()
```
