##### Q. 4) Show all the records where 'No. of Crimes' is 0. And, how many such records are there ?

```python
df[df.no_of_crimes == 0]
```

##### Q. 5) What is the maximum & minimum 'average_price' per year in england ?

```python
# Maximum average Price

df.groupby(by = df.date.dt.year).average_price.max().sort_values(ascending = False).head(1)
```

```python
# Minimum Average Price

df.groupby(by = df.date.dt.year).average_price.min().sort_values(ascending = True).head(1)
```

##### Q. 6) What is the Maximum & Minimum No. of Crimes recorded per area ?

```python
# Maximum No_of_crimes

df.groupby(by = 'area').no_of_crimes.max().sort_values(ascending = False).head(1)
```

```python
# Minimum No_of_crimes

df.groupby(by = 'area').no_of_crimes.min().sort_values(ascending = True).head(1)
```
