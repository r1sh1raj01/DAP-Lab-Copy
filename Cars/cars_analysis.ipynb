```python
import pandas as pd

df = pd.read_csv(r"cars.csv");
print(df);
```

```python
df.isnull().sum()



# print(my_df['Age'].mean())
```

```python
df['Cylinders'].fillna(df['Cylinders'].mean(), inplace=True)
```

```python
df.fillna(df['Cylinders'].mean(), inplace=True)
df.isnull().sum()
```

```python
df['Make'].value_counts()
```

```python
df[df['Origin'].isin(["Asia","Europe"])]
```

```python
df[-(df['Weight'] > 4000)]
```

```python
df['MPG_City'] = df['MPG_City'].apply(lambda x:x+3)
```
